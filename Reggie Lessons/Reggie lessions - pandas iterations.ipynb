```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.DataFrame(np.arange(1, 101).reshape(10,10))
```

```python
df
```

```python
df.values
```

## The below shows that df.values is a list of lists, so i is a list

```python
#This shows that df.values is a list of lists, so i is a list
for i in df.values:
    print(i)
```

## The below shows how to print the values with bracket notation

```python
#This shows how to print the values with bracket notation
for i in df.values:
    print(i[2])
```

## By using the enumerate function, I get an integer for the row number

```python
#By using the enumerate function, I get an integer for the row number
for i, j in enumerate(df.values):
    print(i, j[i])
```

## Assigning a value to the dataframe using df.iat[ ]

```python
for i, j in enumerate(df.values):
    df.iat[i, 4] = 999
df
```

## Using an if statement and numpy assignment types

```python
for i in df.values:
    if i[7]>50:
        i[8] = np.sqrt(i[8])
    #print(i[3])
df
```

## This shows the structure of the namedtuples that itertuples() generates

```python
for row in df.itertuples():
    print(row)
```

## This shows how to access the index integers when using itertuples()

```python
for row in df.itertuples():
    print(row.Index)
```

## Changing column names to distinguish between values and column names

```python
df.columns = ['zero', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']
df
```

## Setting a value using the namedtuples in itertuples()

```python
for row in df.itertuples():
    if row.Index == 0:
        df.at[row.Index,'one'] = 999
df
```

## Using enumerate with itertuples() to access and change a cell value

```python
for i, row in enumerate(df.itertuples()):
    if i == 5:
        df.at[row.Index, 'five'] = 5555
df
```
