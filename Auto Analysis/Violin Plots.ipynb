```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
spx = pd.read_csv('SP500-analysis-daily.csv', index_col='Date', parse_dates=True)
```

```python
sns.violinplot(x='Market Direction',y='Daily Chg Pct', data=spx);
```

```python
sns.violinplot(x='Market Direction',y='ATR20 Pct', data=spx);
```

```python
spx['Market Type'] = spx['Market Direction'] + ' - ' + spx['Market Volatility']
```

```python
sns.violinplot(y='Market Type', x='Daily Chg Pct', data=spx, scale='count');
```

```python
sns.violinplot(y='Market Type', x='ATR20 Pct', data = spx, width=0.8);
```
